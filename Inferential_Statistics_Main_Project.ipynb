## **Problem 2: Breaking Strength of Gunny Bags**

```python
# Analyzing the distribution of breaking strength for cement packaging gunny bags.

# - Mean (μ): 5 kg/cm²
# - Standard Deviation (σ): 1.5 kg/cm²
# - Distribution Assumed: Normal
```

### **2.1 What proportion of the gunny bags have a breaking strength of less than 3.17 kg per sq cm?**

### Calculating the Probability of Breaking Strength < 3.17 kg/cm²

In this section, we calculate the **Z-score** to determine how far 3.17 kg/cm² is from the mean breaking strength. Then, we use the **cumulative distribution function (CDF)** of the standard normal distribution to compute the probability that a randomly selected gunny bag has a breaking strength **less than 3.17 kg/cm²**.


```python
from scipy.stats import norm

mu = 5
sigma = 1.5
x = 3.17

z = (x - mu) / sigma
prob = norm.cdf(z)

print(f"Z-score: {z}")
print(f"P(X > 8) = {prob:.4f} or {prob*100:.2f}%")
```

```python
x_vals = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y_vals = norm.pdf(x_vals, loc=mu, scale=sigma)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='Normal Distribution', color='blue')
plt.fill_between(x_vals, y_vals, 0, where=(x_vals < x), color='skyblue', alpha=0.6)

plt.axvline(x, color='red', linestyle='--', label=f'X = {x}')
plt.title('Probability that Breaking Strength < 3.17 kg/cm²')
plt.xlabel('Breaking Strength (kg/cm²)')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)

# Add figure caption
plt.figtext(0.5, -0.05, 'Figure 2.1: Area under the curve for P(X < 3.17)', wrap=True,
            horizontalalignment='center', fontsize=10)

plt.show()
```

### **2.2 What proportion of the gunny bags have a breaking strength of at least 3.6 kg per sq cm.?**

### Calculating the Probability of Breaking Strength ≥ 3.6 kg/cm²

In this section, we calculate the **Z-score** for 3.6 kg/cm² and determine the **right-tail probability**, i.e., the proportion of gunny bags with a breaking strength **greater than or equal to 3.6 kg/cm²**.  
We use **1 - CDF(z)** to obtain this probability.

The shaded area **to the right of X = 3.6** under the normal curve represents this probability.


```python
from scipy.stats import norm

mu = 5
sigma = 1.5
x = 3.6

z = (x - mu) / sigma
prob = 1-norm.cdf(z)

print(f"Z-score: {z}")
print(f"P(X >= 3.6) = {prob:.4f} or {prob*100:.2f}%")
```

```python
x_vals = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y_vals = norm.pdf(x_vals, loc=mu, scale=sigma)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='Normal Distribution', color='blue')
plt.fill_between(x_vals, y_vals, 0, where=(x_vals >= x), color='skyblue', alpha=0.6)

plt.axvline(x, color='red', linestyle='--', label=f'X = {x}')
plt.title('Probability that Breaking Strength ≥ 3.6 kg/cm²')
plt.xlabel('Breaking Strength (kg/cm²)')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)

# figure caption
plt.figtext(0.5, -0.05, 'Figure 2.2: Area under the curve for P(X ≥ 3.6)', wrap=True,
            horizontalalignment='center', fontsize=10)

plt.show()
```

### **2.3 What proportion of the gunny bags have a breaking strength between 5 and 5.5 kg per sq cm.?**

###  Proportion of Gunny Bags with Breaking Strength Between 5 and 5.5 kg/cm²

In this part, we compute the probability that the breaking strength **falls between 5 kg/cm² and 5.5 kg/cm²**.


#### Z-scores:
- **Lower bound (5)**
- **Upper bound (5.5)**


We use **norm.cdf()** to find the cumulative probabilities for both bounds and subtract to get the probability


```python
from scipy.stats import norm

mu = 5
sigma = 1.5

x1 = 5
x2 = 5.5

p1 = norm.cdf(x1, loc=mu, scale=sigma)
p2 = norm.cdf(x2, loc=mu, scale=sigma)

probability = p2 - p1
print(f"P(5 ≤ X ≤ 5.5) = {probability:.4f} or {probability*100:.2f}%")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mu = 5     # Mean
sigma = 1.5  # Standard Deviation
x1 = 5
x2 = 5.5

# X values and PDF
x_vals = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y_vals = norm.pdf(x_vals, loc=mu, scale=sigma)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='Normal Distribution', color='blue')
plt.fill_between(x_vals, y_vals, 0, where=((x_vals >= x1) & (x_vals <= x2)), color='skyblue', alpha=0.6)
plt.axvline(x1, color='green', linestyle='--', label=f'X = {x1}')
plt.axvline(x2, color='red', linestyle='--', label=f'X = {x2}')
plt.title('Probability that Breaking Strength is Between 5 and 5.5 kg/cm²')
plt.xlabel('Breaking Strength (kg/cm²)')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.figtext(0.5, -0.05, 'Figure 2.3: Area under the curve for P(5 ≤ X ≤ 5.5)', wrap=True,
            horizontalalignment='center', fontsize=10)
plt.tight_layout()
plt.show()
```

### **2.4 What proportion of the gunny bags have a breaking strength NOT between 3 and 7.5 kg per sq cm.?**

### Proportion of Gunny Bags with Breaking Strength **NOT** Between 3 and 7.5 kg/cm²

In this part, we calculate the probability that a gunny bag has a breaking strength **outside the range [3, 7.5] kg/cm²**.  
This is calculated using the **complement rule**:



We find


```python
from scipy.stats import norm

mu = 5
sigma = 1.5

# CDF values
p_low = norm.cdf(3, loc=mu, scale=sigma)
p_high = norm.cdf(7.5, loc=mu, scale=sigma)

# Probability NOT between 3 and 7.5
p_not_between = 1 - (p_high - p_low)
print(f"P(X not between 3 and 7.5) = {p_not_between:.4f}")
```

## **Problem Statement 3: Comparative Analysis of Stone Hardness – Suitability and Treatment Effect**

### **3.1 Zingaro has reason to believe that the unpolished stones may not be suitable for printing. Do you think Zingaro is justified in thinking so?**

Loading the Dataset and Preparing Data  
The dataset contains Brinell Hardness Index (BHI) values for unpolished and polished stones. We will use the unpolished data to test if the stones meet the required threshold of 150 for printing suitability.


```python
df = pd.read_csv("/content/drive/MyDrive/Inferential Statistics/Zingaro_Company.csv")

print(df.head())
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
# Removes leading and trailing whitespace from column names (optional).
# Done this step as there is trailing space after the column name 'Unpolished '

unpolished_data = df['Unpolished']
polished_data = df['Treated and Polished']

unpolished_data
```

Performing One-Sample t-Test for Unpolished Stones  
We conduct a one-sample, one-tailed t-test to test if the mean hardness of unpolished stones is significantly lower than 150.


```python
# Hypothesis test
t_stat, p_value = stats.ttest_1samp(unpolished_data, popmean=150)

# One-tailed test (left side)
p_value_one_tailed = p_value / 2

# Mean calculation
mean_val = unpolished_data.mean()

# Results
print(f"Mean of unpolished stones: {mean_val:.2f}")
print(f"t-statistic = {t_stat:.3f}")
print(f"p-value (one-tailed) = {p_value_one_tailed:.4f}")
```

Figure 3.1: One-sample t-test visualization for unpolished stones. The shaded red area represents the rejection region of the hypothesis. The sample mean clearly lies within this region, supporting rejection of the null hypothesis.


```python
# ---- Plotting the t-distribution ----

# Sample size and statistics
n = len(unpolished_data)
mean_val = unpolished_data.mean()
sample_std = np.std(unpolished_data, ddof=1)


dfree = n - 1
x = np.linspace(mean_val - 4*sample_std, mean_val + 4*sample_std, 500)
t_dist = stats.t(dfree, loc=150, scale=sample_std / np.sqrt(n))
y = t_dist.pdf(x)

# Critical t-value for 5% significance (left-tailed)
critical_value = stats.t.ppf(0.05, dfree, loc=150, scale=sample_std / np.sqrt(n))
critical_x = np.linspace(x[0], critical_value, 500)

plt.figure(figsize=(10, 6))
plt.plot(x, y, color='blue', label='t-Distribution under H₀')

# Shade rejection region
plt.fill_between(critical_x, 0, t_dist.pdf(critical_x), color='red', alpha=0.4, label='Rejection Region (α=0.05)')

# Plot observed sample mean and hypothesized mean
plt.axvline(mean_val, color='black', linestyle='--', label=f'Sample Mean = {mean_val:.2f}')
plt.axvline(150, color='green', linestyle='--', label=f'H₀ Mean = 150')

plt.title("Figure 3.1: One-Sample t-Test Visualization")
plt.xlabel("Brinell Hardness Index (BHI)")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

### **3.2 Is the mean hardness of the polished and unpolished stones the same?**

Performing Welch’s Two-Sample t-Test (Unequal Variance)  
We perform an **independent two-sample t-test (Welch’s t-test)** to determine whether there is a **significant difference in mean hardness** between polished and unpolished stones.


```python
# Performing Welch’s t-test (assumes unequal variances)
t_stat, p_value = stats.ttest_ind(unpolished_data, polished_data, equal_var=False)

# Calculating means for reference
mean_unpolished = unpolished_data.mean()
mean_polished = polished_data.mean()

# Output results
print(f"Mean of unpolished stones: {mean_unpolished:.2f}")
print(f"Mean of polished stones: {mean_polished:.2f}")
print(f"t-statistic = {t_stat:.3f}")
print(f"p-value (two-tailed) = {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis. There is a significant difference in mean hardness.")
else:
    print("Conclusion: Fail to reject the null hypothesis. No significant difference in mean hardness.")
```

Visualizing the Two-Sample t-Test Result  
This plot shows two overlapping normal distributions for **unpolished (red)** and **polished (green)** stones.  
The **horizontal dashed lines** represent the means of each group.  
It visually emphasizes the shift in the distribution of polished stones toward higher Brinell Hardness Index (BHI) values.

## **Problem 4: Multi-Factor Analysis of Dental Implant Hardness**

### **4.1 How does the hardness of implants vary depending on dentists?**

We begin by checking the assumptions of the One-Way ANOVA test for each alloy. Specifically, we test for normality of the data using the Shapiro-Wilk Test and equality of variances using Levene’s Test.

```python
import pandas as pd
from scipy.stats import shapiro, levene, f_oneway
from statsmodels.formula.api import ols
import statsmodels.api as sm

# Load dataset
df = pd.read_excel("/content/drive/MyDrive/Inferential Statistics/Dental Hardness data.xlsx")

# Split data by alloy
df_alloy1 = df[df['Alloy'] == 1]
df_alloy2 = df[df['Alloy'] == 2]

# Grouping data by dentist for each alloy
groups_alloy1 = [group['Response'].values for _, group in df_alloy1.groupby('Dentist')]
groups_alloy2 = [group['Response'].values for _, group in df_alloy2.groupby('Dentist')]

# ---- ALLOY 1 ----
# Combine responses for Shapiro-Wilk Test (can also use residuals later)
combined_alloy1 = pd.concat([pd.Series(g) for g in groups_alloy1])
shapiro_stat1, shapiro_p1 = shapiro(combined_alloy1)
levene_stat1, levene_p1 = levene(*groups_alloy1)

# ---- ALLOY 2 ----
combined_alloy2 = pd.concat([pd.Series(g) for g in groups_alloy2])
shapiro_stat2, shapiro_p2 = shapiro(combined_alloy2)
levene_stat2, levene_p2 = levene(*groups_alloy2)

# ---- Print assumption test results (with statistics) ----
print("===== Assumption Checks =====")
print(f"ALLOY 1: Shapiro-Wilk W = {shapiro_stat1:.4f}, p = {shapiro_p1:.4f}")
print(f"ALLOY 1: Levene’s F = {levene_stat1:.4f}, p = {levene_p1:.4f}")
print(f"ALLOY 2: Shapiro-Wilk W = {shapiro_stat2:.4f}, p = {shapiro_p2:.4f}")
print(f"ALLOY 2: Levene’s F = {levene_stat2:.4f}, p = {levene_p2:.4f}")
```

The following Q-Q plots visually assess the normality of residuals for each alloy. Deviations from the diagonal line indicate departures from normality.

* Figure 4.1.1: Q-Q Plot of Residuals for Alloy 1
* Figure 4.1.2: Q-Q Plot of Residuals for Alloy 2

```python
for i, alloy in enumerate(df['Alloy'].unique(), start=1):
    alloy_df = df[df['Alloy'] == alloy]
    model = ols('Response ~ C(Dentist)', data=alloy_df).fit()

    sm.qqplot(model.resid, line='45')
    plt.title(f"Figure 4.1.{i}: Q-Q Plot of Residuals (Alloy {alloy})")
    plt.show()
```

After confirming assumptions, we conduct One-Way ANOVA to determine whether the mean implant hardness significantly differs across dentists for each alloy.

```python
# ---- Fit ANOVA model ----
model1 = ols('Response ~ C(Dentist)', data=df_alloy1).fit()
model2 = ols('Response ~ C(Dentist)', data=df_alloy2).fit()

anova1 = sm.stats.anova_lm(model1, typ=2)
anova2 = sm.stats.anova_lm(model2, typ=2)

print("\nAlloy 1 ANOVA Table:\n", anova1)
print("\nAlloy 2 ANOVA Table:\n", anova2)
```

### **4.2 How does the hardness of implants vary depending on methods?**

We begin by checking the assumptions of normality and homogeneity of variances for each alloy before performing One-Way ANOVA.

```python
import pandas as pd
from scipy.stats import shapiro, levene
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_excel("/content/drive/MyDrive/Inferential Statistics/Dental Hardness data.xlsx")

# Function to perform complete analysis for each alloy
def analyze_by_method(alloy_number):
    print(f"\n===== Analysis for Alloy {alloy_number} =====")

    df_alloy = df[df['Alloy'] == alloy_number]

    # Assumptions Check BEFORE ANOVA

    # 1️ Normality (Shapiro-Wilk) → Applied to Response for simplicity
    combined_responses = df_alloy['Response']
    shapiro_stat, shapiro_p = shapiro(combined_responses)

    # 2️ Equal Variance (Levene’s Test) → Compare groups by Method
    groups = [group['Response'].values for _, group in df_alloy.groupby("Method")]
    levene_stat, levene_p = levene(*groups)

    print(f"Shapiro-Wilk Test: W = {shapiro_stat:.4f}, p = {shapiro_p:.4f}")
    print(f"Levene's Test:    F = {levene_stat:.4f}, p = {levene_p:.4f}")

analyze_by_method(1)
analyze_by_method(2)
```

ANOVA Table

```python
model1 = ols('Response ~ C(Method)', data=df_alloy1).fit()
model2 = ols('Response ~ C(Method)', data=df_alloy2).fit()

anova1 = sm.stats.anova_lm(model1, typ=2)
anova2 = sm.stats.anova_lm(model2, typ=2)

print("\nAlloy 1 ANOVA Table:\n", anova1)
print("\nAlloy 2 ANOVA Table:\n", anova2)
```

After confirming that the ANOVA shows significant differences (p < 0.05), we conduct Tukey’s HSD test to identify which specific implant methods differ in hardness.

```python
def tukey_posthoc_by_method(alloy_number):
    print(f"\n===== Tukey HSD Post Hoc Test for Alloy {alloy_number} =====")
    df_alloy = df[df['Alloy'] == alloy_number]

    # Fit ANOVA model for completeness
    model = ols('Response ~ C(Method)', data=df_alloy).fit()
    anova = sm.stats.anova_lm(model, typ=2)
    print("\nANOVA Table:")
    print(anova)

    # Conduct Tukey HSD (only if p < 0.05)
    if anova['PR(>F)'].iloc[0] < 0.05:
        tukey = pairwise_tukeyhsd(endog=df_alloy['Response'], groups=df_alloy['Method'])
        print("\nTukey HSD Results:")
        print(tukey.summary())

        # Plot the Tukey HSD result
        tukey.plot_simultaneous()
        plt.title(f'Tukey HSD (Alloy {alloy_number})')
        plt.xlabel('Mean Difference')
        plt.show()
    else:
        print("Post hoc test not required as ANOVA result is not significant (p > 0.05).")


# Run Tukey HSD for both alloys
tukey_posthoc_by_method(1)
tukey_posthoc_by_method(2)
```

Boxplots help visualize the spread and central tendency of implant hardness for each method and highlight differences revealed by Tukey’s test.

```python
def boxplot_by_method(alloy_number):
    df_alloy = df[df['Alloy'] == alloy_number]

    plt.figure(figsize=(8, 5))
    sns.boxplot(x='Method', y='Response', data=df_alloy, hue='Method', palette="Set2", legend=False)
    plt.title(f'Boxplot of Implant Hardness by Method (Alloy {alloy_number})')
    plt.xlabel('Implant Method')
    plt.ylabel('Implant Hardness')
    plt.show()

boxplot_by_method(1)
boxplot_by_method(2)
```

### **4.4 How does the hardness of implants vary depending on dentists and methods together?**

Perform Shapiro-Wilk and Levene’s tests to check normality of residuals and homogeneity of variances before conducting Two-Way ANOVA.


```python
import pandas as pd
from scipy.stats import shapiro, levene
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_excel("/content/drive/MyDrive/Inferential Statistics/Dental Hardness data.xlsx")

# Function to perform assumptions check for each alloy
def check_assumptions_two_way(alloy_number):
    print(f"\n===== Assumption Checks for Alloy {alloy_number} =====")

    df_alloy = df[df['Alloy'] == alloy_number]

    # Fit Two-Way ANOVA model with interaction to get residuals
    model = ols('Response ~ C(Dentist) * C(Method)', data=df_alloy).fit()
    residuals = model.resid

    # Normality → Shapiro-Wilk Test on residuals of the full model
    shapiro_stat, shapiro_p = shapiro(residuals)

    # Equal Variance → Levene’s Test on combinations of Dentist and Method
    groups = [group['Response'].values for _, group in df_alloy.groupby(['Dentist', 'Method'])]
    levene_stat, levene_p = levene(*groups)

    print(f"Shapiro-Wilk Test (Normality of residuals): p = {shapiro_p:.4f}")
    print(f"Levene’s Test (Homogeneity of variances): p = {levene_p:.4f}")

# Run for both alloys
check_assumptions_two_way(1)
check_assumptions_two_way(2)
```

Conducting Two-Way ANOVA to assess the individual and combined effects of dentist and method on implant hardness for each alloy.


```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Load dataset
df = pd.read_excel("/content/drive/MyDrive/Inferential Statistics/Dental Hardness data.xlsx")

# Prepare data for Alloy 1 and Alloy 2
df_alloy1 = df[df['Alloy'] == 1]
df_alloy2 = df[df['Alloy'] == 2]

def display_anova_results(model, name):
    anova = sm.stats.anova_lm(model, typ=2)
    anova['mean_sq'] = anova['sum_sq'] / anova['df']
    print(f"\nANOVA Table: {name}")
    print(anova[['sum_sq', 'df', 'mean_sq', 'F', 'PR(>F)']])

# ---- For Alloy 1 (Two-Way ANOVA with Interaction) ----
model1_full = ols('Response ~ C(Dentist) * C(Method)', data=df_alloy1).fit()
display_anova_results(model1_full, 'Alloy 1 (Two-Way with Interaction)')

# ---- For Alloy 2 (Two-Way ANOVA with Interaction) ----
model2_full = ols('Response ~ C(Dentist) * C(Method)', data=df_alloy2).fit()
display_anova_results(model2_full, 'Alloy 2 (Two-Way with Interaction)')
```

```python
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols

# ---- For Alloy 1 ----
model_alloy1 = ols('Response ~ C(Dentist) * C(Method)', data=df_alloy1).fit()

plt.figure(figsize=(8, 5))
plt.scatter(model_alloy1.fittedvalues, model_alloy1.resid, alpha=0.7)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel("Fitted Values")
plt.ylabel("Residuals")
plt.title("Residual Plot (Alloy 1)")
plt.show()

# ---- For Alloy 2 ----
model_alloy2 = ols('Response ~ C(Dentist) * C(Method)', data=df_alloy2).fit()

plt.figure(figsize=(8, 5))
plt.scatter(model_alloy2.fittedvalues, model_alloy2.resid, alpha=0.7)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel("Fitted Values")
plt.ylabel("Residuals")
plt.title("Residual Plot (Alloy 2)")
plt.show()
```

Performing Tukey HSD post hoc analysis to identify which implant methods significantly differ in their effect on implant hardness.


```python
import pandas as pd
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Load dataset
df = pd.read_excel("/content/drive/MyDrive/Inferential Statistics/Dental Hardness data.xlsx")

# Prepare data for Alloy 1 and Alloy 2
df_alloy1 = df[df['Alloy'] == 1]
df_alloy2 = df[df['Alloy'] == 2]

# ---- Tukey HSD for Method → Alloy 1 ----
tukey1 = pairwise_tukeyhsd(endog=df_alloy1['Response'],
                           groups=df_alloy1['Method'],
                           alpha=0.05)
print("Tukey HSD Results for Method (Alloy 1):")
print(tukey1.summary())

# ---- Tukey HSD for Method → Alloy 2 ----
tukey2 = pairwise_tukeyhsd(endog=df_alloy2['Response'],
                           groups=df_alloy2['Method'],
                           alpha=0.05)
print("\nTukey HSD Results for Method (Alloy 2):")
print(tukey2.summary())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.factorplots import interaction_plot

# ---- Make clean copy and reset index → avoids KeyError ----
df_alloy1_plot = df_alloy1.copy().reset_index(drop=True)
df_alloy2_plot = df_alloy2.copy().reset_index(drop=True)

# Convert columns to STRING as required by interaction_plot()
df_alloy1_plot['Method'] = df_alloy1_plot['Method'].astype(str)
df_alloy1_plot['Dentist'] = df_alloy1_plot['Dentist'].astype(str)
df_alloy2_plot['Method'] = df_alloy2_plot['Method'].astype(str)
df_alloy2_plot['Dentist'] = df_alloy2_plot['Dentist'].astype(str)

# Colors/Markers
colors1 = sns.color_palette("tab10", n_colors=df_alloy1_plot['Dentist'].nunique())
colors2 = sns.color_palette("tab10", n_colors=df_alloy2_plot['Dentist'].nunique())
markers = ['o', 's', '^', 'D', 'v']

# ---- Interaction Plot → Alloy 1 ----
plt.figure(figsize=(8, 5))
interaction_plot(df_alloy1_plot['Method'], df_alloy1_plot['Dentist'], df_alloy1_plot['Response'],
                 markers=markers[:df_alloy1_plot['Dentist'].nunique()], colors=colors1)
plt.title("Interaction Plot: Dentist vs Method (Alloy 1)")
plt.xlabel("Implant Method")
plt.ylabel("Implant Hardness")
plt.show()

# ---- Interaction Plot → Alloy 2 ----
plt.figure(figsize=(8, 5))
interaction_plot(df_alloy2_plot['Method'], df_alloy2_plot['Dentist'], df_alloy2_plot['Response'],
                 markers=markers[:df_alloy2_plot['Dentist'].nunique()], colors=colors2)
plt.title("Interaction Plot: Dentist vs Method (Alloy 2)")
plt.xlabel("Implant Method")
plt.ylabel("Implant Hardness")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="Method", y="Response", hue="Dentist", data=df_alloy1_plot, palette="Set2")
plt.title("Boxplot of Implant Hardness by Method and Dentist (Alloy 1)")
plt.xlabel("Implant Method")
plt.ylabel("Implant Hardness")
plt.legend(title='Dentist')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x="Method", y="Response", hue="Dentist", data=df_alloy2_plot, palette="Set2")
plt.title("Boxplot of Implant Hardness by Method and Dentist (Alloy 2)")
plt.xlabel("Implant Method")
plt.ylabel("Implant Hardness")
plt.legend(title='Dentist')
plt.show()
```
